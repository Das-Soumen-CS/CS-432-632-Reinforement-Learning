```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# squarrel_pos = state
# variables
action_space = 2
num_states = 12

reward_table = np.zeros([num_states, action_space]) 
print("intial reward table=\n",reward_table)
gamma = 0.9 #discount rate
num_of_eps = 200

# environment
env = np.array([-5, -1, -1, -1, 10, -1, -1, -1, -1, -1, -1, -5]) 
nut_pos = 4
end_pos_1 = 0
end_pos_2 = 11
sleeping = False
    
# store scores
scores = []    

# Store the number of extra steps
number_of_extra_steps = []

# define step function
def step(pos, value):
        
        # determine action
        if value == 0:
            next_state = pos - 1 #left
        
        if value == 1:
            next_state = pos + 1 #right
            
        # verify action 
        if next_state > end_pos_2:
            next_state = end_pos_2
            
        if next_state < end_pos_1:
            next_state = end_pos_1
        
        # reward
        reward = env[next_state]
        
        # return
        return next_state, reward

# training loop
for ep in range(1, num_of_eps):

    # print episode
    print("Episode: ", ep)
    
    # recreate environment
    env = np.array([-5, -1, -1, -1, 10, -1, -1, -1, -1, -1, -1, -5])
    
    # choose a random position for squarrel
    squarrel_pos = random.choice(range(end_pos_1+1, end_pos_2+1))
    
    # be sure that squarrel is not on the nut
    while squarrel_pos == nut_pos:
        squarrel_pos = random.choice(range(1, 10))
        
        if squarrel_pos != nut_pos:
            break
    
    # find the distance between nut and squarrel
    distance = np.absolute(squarrel_pos - nut_pos)
    
    # print squarrel position
    print("Squarrel Position: ", squarrel_pos)
    
    # place a zero to represent the squarrel
    env[squarrel_pos] = 0
    
    # print the first state of the environment
    print("Environment:", env, "\n----------------------------------------------------")
    
    # store rewards
    rewards = [] 
    
    number_of_steps = 0
    
    # game loop
    while squarrel_pos != nut_pos:
        
        # find which number is squarrel on
        if (squarrel_pos == end_pos_1) or (squarrel_pos == end_pos_2):
            last_num = -5
            
        if (squarrel_pos != end_pos_1) and (squarrel_pos != end_pos_2):
            last_num = -1
        #print("squarrel_pos=",squarrel_pos)
        # Place back the number that squarrel was blocking
        env[squarrel_pos] = last_num
        #print("env[squarrel_pos]=",env[squarrel_pos])
        # max action in the table
        #print(reward_table[squarrel_pos])
        action = np.argmax(reward_table[squarrel_pos])
        #print("action=",action)
        # take action   "function call"  
        next_state, reward = step(squarrel_pos, action)
        
        # append reward to rewards list
        rewards.append(reward)
        
        # bounds of the environment
        if squarrel_pos < end_pos_1:
            squarrel_pos = end_pos_1
            
        if squarrel_pos > end_pos_2:
            squarrel_pos = end_pos_2
        
        # Update reward table with bellman equation
        reward_table[squarrel_pos][action] = reward + gamma * max(reward_table[next_state])
        
        # update state
        squarrel_pos = next_state
        
        # show squarrel's position
        env[squarrel_pos] = 0
        
        # show environment
        print(env)
            
        # pause for 1.2 seconds so we can see what is happening
        if sleeping:
            time.sleep(1.2) 
        
        # increase number_of_steps by 1
        number_of_steps += 1
        
    # find total score by summing the rewards
    total_score = sum(rewards)
    
    # append total score to scores list
    scores.append(total_score)
    
    # print final score of the episode
    print("Reward_Score:", total_score)
    
    # find the number of extra steps 
    extra_steps = np.absolute(number_of_steps - distance)
    
    # append number of extra steps to list
    number_of_extra_steps.append(extra_steps)
    
    # print how many extra steps the agent has took
    print("Number of Extra Steps:", extra_steps)
    
average_score = np.mean(scores)
print("Training Completed. \nAverage Score: {}".format(average_score))
```
