```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
# parameters
action_space = 4
num_states = 25
reward_table = np.zeros([num_states, action_space]) 
gamma = 0.9 #discount rate
num_of_eps = 20
epsilon = 0.1

# environment
"""
-5, -1, -1, -1, -1
-1, -1, +20, -1, -1
-1, -1, -1, -5, -1
-1, -5, -1, -1, -1
-1, -1, -1, -1, -1

"""
env = np.array([[-5,-1,-1,-1,-1], 
                [-1,-1,+20,-1,-1], 
                [-1,-1,-1,-5,-1], 
                [-1,-5,-1,-1,-1], 
                [-1,-1,-1,-1,-1]])

# parameters
nut_pos = [1, 2]
danger_1_pos = [0,0]
danger_2_pos = [2,3]
danger_3_pos = [3,1]

# environment's bounds
end_column_1 = 0
end_column_2 = 4
end_row_1 = 0
end_row_2 = 4

sleeping = False

# scores
scores = []    

number_of_extra_steps = []

# define step function
def step(squarrel_pos_row, squarrel_pos_column, action):
    
    if action == 0:
        new_pos = [squarrel_pos_row, squarrel_pos_column - 1] # left
    
    if action == 1:
        new_pos = [squarrel_pos_row, squarrel_pos_column + 1] # right
    
    if action == 2:
        new_pos = [squarrel_pos_row - 1, squarrel_pos_column] # up
    
    if action == 3:
        new_pos = [squarrel_pos_row + 1, squarrel_pos_column] # down
        
    # verify action
    if new_pos[0] < end_row_1:
        new_pos[0] = end_row_1
            
    if new_pos[0] > end_row_2:
        new_pos[0] = end_row_2
            
    if new_pos[1] < end_column_1:
        new_pos[1] = end_column_1
            
    if new_pos[1] > end_column_2:
        new_pos[1] = end_column_2
        
    # define reward
    reward = env[new_pos[0], new_pos[1]]
        
    # define next_state
    next_state = new_pos[0]*5 + new_pos[1]
        
    return next_state, reward, new_pos
    
# define random action
def random_action():
    
    # choose a random action
    return random.choice(range(0,4))

# training loop
for ep in range(1, num_of_eps):

    # print episode's number
    print("Episode: ", ep)
    
    # redefine environment
    env = np.array([[-5,-1,-1,-1,-1], 
                    [-1,-1,+20,-1,-1], 
                    [-1,-1,-1,-5,-1], 
                    [-1,-5,-1,-1,-1], 
                    [-1,-1,-1,-1,-1]])
    
    # choose a position for squarrel
    squarrel_pos_column = random.choice(range(end_column_1, end_column_2+1)) 
    squarrel_pos_row = random.choice(range(end_row_1, end_row_2+1))
    
    # verify squarrel's position: it should not be on the nut and predators (danger)
    while ([squarrel_pos_row, squarrel_pos_column] == nut_pos) or ([squarrel_pos_row, squarrel_pos_column] == danger_1_pos) or ([squarrel_pos_row, squarrel_pos_column] == danger_2_pos) or ([squarrel_pos_row, squarrel_pos_column] == danger_3_pos):
    
        # choose new positions for the squarrel
        squarrel_pos_column = random.choice(range(end_column_1, end_column_2+1))
        squarrel_pos_row = random.choice(range(end_row_1, end_row_2+1))

        # if squarrel's position is decent break the loop
        if (squarrel_pos != nut_pos) and (squarrel_pos != danger_1_pos) and (squarrel_pos != danger_2_pos) and (squarrel_pos != danger_3_pos):
            break
            
    squarrel_pos = [squarrel_pos_row, squarrel_pos_column]
    
    # find distance between squarrel and nut
    distance = np.absolute(squarrel_pos[0] - nut_pos[0]) + np.absolute(squarrel_pos[1] - nut_pos[1])
    
    print("Squarrel Position: ", squarrel_pos)
    
    # show squarrel's position
    env[squarrel_pos[0], squarrel_pos[1]] = 0
    
    print("Environment:\n", env, "\n---------------------------------------")
    
    # store rewards
    rewards = [] 
    
    # number of total steps that squarrel has took
    number_of_steps = 0
    
    # game loop
    while squarrel_pos != nut_pos:
        
        # define state
        state = squarrel_pos[0]*5 + squarrel_pos[1]
        
        # find which number is squarrel on
        if (squarrel_pos == danger_1_pos) or (squarrel_pos == danger_2_pos) or (squarrel_pos == danger_3_pos):
            last_num = -5
            
        if (squarrel_pos != danger_1_pos) and (squarrel_pos != danger_2_pos) and (squarrel_pos != danger_3_pos):
            last_num = -1
        
        # place back the number that squarrel was blocking
        env[squarrel_pos[0], squarrel_pos[1]] = last_num
        
        # epsilon greedy
        if np.random.uniform() < epsilon:
            action = random_action()
            print("Random Action:")
            
        else:
            # max action in the table
            action = np.argmax(reward_table[state])

        #take action
        next_state, reward, squarrel_pos = step(squarrel_pos[0], squarrel_pos[1], action)
        
        # store rewards
        rewards.append(reward)
        
        # verify squarrel's position (bounds of the environment)
        if squarrel_pos[0] < end_row_1:
            squarrel_pos[0] = end_row_1
            
        if squarrel_pos[0] > end_row_2:
            squarrel_pos[0] = end_row_2
            
        if squarrel_pos[1] < end_column_1:
            squarrel_pos[1] = end_column_1
            
        if squarrel_pos[1] > end_column_2:
            squarrel_pos[1] = end_column_2
           
        # bellman equation
        reward_table[state][action] = reward + gamma * max(reward_table[next_state])
          
        # show squarrel's position
        env[squarrel_pos[0], squarrel_pos[1]] = 0
        
        # show environment
        print(env)
                
        if sleeping:
            time.sleep(1.2) 
            
        number_of_steps += 1
          
    # calculate total score for the episode
    total_score = sum(rewards)
    
    scores.append(total_score)
    
    print("Score:", total_score)
    
    # find extra steps
    extra_steps = np.absolute(number_of_steps - distance)
        
    number_of_extra_steps.append(extra_steps)
    
    print("Number of Extra Steps:", extra_steps)
    
# print information about the training
average_score = np.mean(scores)
print("Training Completed. \nAverage Score: {}".format(average_score))
```
