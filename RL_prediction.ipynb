```python
import numpy as np
import pandas as pd

''' X_t+1 =A* X_t , A= Two state Transition matrix, where t=0,1,2,3,4
  want to predict at time (t+1) based on the observation at time t'''

#P_move(To city /From city) =0.95 ,
#P_move(To city /From Suburbs) =0.03 ,
#P_move(To Suburb /From city) =0.05 ,
#P_move(To Suburbs From Suburbs) =0.95 

A=np.array([[0.95 ,0.3],[0.05 ,0.97]]) # A is the transition probabiity matrix
print("Transition probability matrix =\n",A)

# city population in 2000 is =6,000000 ,Suburbs population in 2000 is =4,00000
c_pop_2000=600000
Sub_pop_2000=400000

#Probability distribution=> PD
PD_city= c_pop_2000/(c_pop_2000+Sub_pop_2000)
PD_Sub=Sub_pop_2000/(c_pop_2000+Sub_pop_2000)

x_0 = np.array([PD_city,PD_Sub])

print("\nProbaility Distribuiton=\n",x_0,"\n")

# Matrix multiplication without inbuild function
Result=np.matmul(A,x_0)
steps=int(input("Enter the iteration no="))
print("Itreation_No_0=",Result)
for i in range(1,steps):
    Result =np.matmul(A,Result)
    print("Itreation_No_",i,"=",Result)

# As for example if I want to compute X_t+1 =A* X_t  whent t=19 ,Hence X_20= A^20 *X_0   , With inbulid library
A_steps = np.linalg.matrix_power(A, steps)

# compute x^n
x_steps = A_steps @ x_0
print("\n Without Library Compute x^n= ",x_steps)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from colorama import Fore, Back, Style
from termcolor import colored
import simple_colors

####REINFORCEMENT LEARNING ITERATIVE Q_TABLE PREPARE:####

''' X_t+1 =A* X_t , A= Two state Transition matrix, where t=0,1,2,3,4
  want to predict at time (t+1) based on the observation at time t'''

A=np.array([[.5,.2,.3],[.3,.8,.3],[.2, 0,.4]]) # A is the transition probabiity matrix
print("Transition probability matrix =\n",A)

# city population in 2000 is =6,000000 ,Suburbs population in 2000 is =4,00000
C_pop_2000=600000
Sub_pop_2000=300000
Vil_pop_2000=100000

#Probability distribution=> PD
PD_city=C_pop_2000/(C_pop_2000+Sub_pop_2000+Vil_pop_2000)
PD_Sub=Sub_pop_2000/(C_pop_2000+Sub_pop_2000+Vil_pop_2000)
PD_Vil=Vil_pop_2000/(C_pop_2000+Sub_pop_2000+Vil_pop_2000)

x_0 = np.array([PD_city,PD_Sub,PD_Vil])

print("\nProbaility Distribuiton=\n",x_0,"\n")

# Matrix multiplication without inbuild function
Result=np.matmul(A,x_0)
List=[]

steps=int(input(Fore.GREEN +"Enter the iteration no="))
print("\n")
np.set_printoptions(formatter={'float': lambda x: "{0:0.3f}".format(x)})
print("\033[1;31m ","Itreation_No_0=",Result)
for i in range(1,steps):
    Result =np.matmul(A,Result)  
    #print("Itreation_No_", i,"=", Result)  '''It will not provide an exact approximate  to see; please uncomment it'''
    X=Result[0]
    Y=Result[1]
    Z=Result[2]
    temp=[X,Y,Z]
    if((i-1)%2==0):
        print("\033[1;34m ","Itreation_No_",i,"=",temp)
    else:
        print("\033[1;31m ","Itreation_No_",i,"=",temp)

    list=Result.tolist()
    List.append(list)
    if(i>=2):
        #print(List[i-2],List[i-1])
        if (List[i-2]==List[i-1]):
            #print(Fore.GREEN +"\nStopping criteria!= Last 2 consecutive operations are exact same\n")
            print(simple_colors.green("\nStopping criteria!= Last 2 consecutive operations are exact same\n", 'bold'))
            print("\033[1;34m ",List[i-2],"\033[1;31m",List[i-1])
            print(Style.RESET_ALL)
            break
                  
#print("Hi",len(List))
print("Append the convergence value in a List For Future use :\n",List)
   
```
